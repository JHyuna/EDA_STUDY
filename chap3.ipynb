### pandas의 특수한 범주형 데이터 형식으로 변경해 메모리 사용량 획기적 감소

```python
col2['STABBR'] = col2['STABBR'].astype('category')
col2.dtypes
```

```python
new_mem = col2.memory_usage(deep=True)
new_mem
```

### 상위~ 하위~ (최대/최소) 선택

```python
movie = pd.read_csv('movie.csv')
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.head()
```

```python
#nlargest method
movie2.nlargest(100, 'imdb_score').head() #평점상위100개
```

```python
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget') #평점 상위 100개 중에 최저에산 5개
```

nlargest, nsmallest에서 열이름 list로 전달하면 첫번째 열에서 동률이 있으면 그 다음 열을 사용해 순서를 결정한다.  

```python
movie3 = movie[['movie_title', 'title_year', 'imdb_score']]
```

```python
#sort_values메소드
movie3.sort_values('title_year', ascending=False).head()
#ascending default는 오름차순. False는 내림차순
```

```python
#여러열 동시 정렬

movie3 = movie3.sort_values(['title_year','imdb_score'], ascending=False)
movie3.head()
```

```python
#drop_duplicates method를 사용해 각 연도 첫 행만 남기고 없애기
movie_top_year = movie3.drop_duplicates(subset='title_year')
movie_top_year.head()
```

drop_duplicates 는 처음 등장하는 고유한 행만 유지하고 나머지 제거한다. 처음에는 모든 행이 고유한 값이므로 아무런 행도 제거되지 않지만 subset에 값을 지정해주면 고유한 값은 오직 지정한 열들에 대해서만 고려되기 때문에 title_year 기준으로 각 연도 하나의 행만 반환  
keep parameter을 last로 하면 각 그룹 마지막 행 선택, False로 설정하면 모든 중복값은 완전 삭제

```python
#하나는 오름/하나는 내림차순 정렬해보기
movie4 = movie[['movie_title', 'title_year', 'content_rating', 'budget']]
movie4_sorted = movie4.sort_values(['title_year', 'content_rating', 'budget'], ascending=[False, False, True])
movie4_sorted.head()
```

### Trailing stop order price

주식거래에 사용할 수 있는 전략 중 하나. 지정가 주문 stop order  
투자가가 주가가 특정 지점에 도달하면 매입이나 매도 주문 내는 것  
대규모 손실을 막거나 이익 보존에 유용  
연속적으로  지정가 주문의 판매가를 수정해 주식 가치가 상승할 경우 이것을 따라 가는 것이 Trailing stop order price 추적 지정가 주문  
뭐 예를 들어 주식이 원래 100$에서 120$로 상승하면 10% 낮은 현재 시장가는 108 달러 원래는 90 달러에서 변경 된 것  
매입 후 최대가에 연동돼 있는 것  
주식이 120달러에서 110달러로 떨어져도 여전히 추적 지정가 주문은 108달러

```python
import pandas_datareader as pdr 
# 주가 정보를 온라인으로 읽어오는 제삼자패키지
```

```python
tsla = pdr.DataReader('tsla', data_source='yahoo', start='2017-1-1')
tsla.head(8)
```

```python
tsla_close = tsla['Close']
```

```python
#cummax method: 최근 값까지 포함해 그 중 가장 최댓값 찾는다.이 열에 0.9나 다른 손실제한 값 곱하면 추적 지정가 주문 만들 수 있음.
tsla_cummax = tsla_close.cummax()
tsla_cummax.head(10)
```

```python
tsla_trailing_stop = tsla_cummax * .9
tsla_trailing_stop.head(8)
```

실제 지정가 주문 실행시 수익이 얼마인지 계산해보기  

```python
def set_trailing_loss(symbol, purchase_date, perc): #종목코드 , 구매일, 지정가퍼센트
    close = pdr.DataReader(symbol, 'yahoo', start=purchase_date)['Close']
    return close.cummax()*perc
```

```python
msft_trailing_stop = set_trailing_loss('msft', '2017-6-1', .85)
msft_trailing_stop
```
